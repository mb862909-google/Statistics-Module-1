# **Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.**

Answer: Here‚Äôs the complete code with explanation and output ‚Äî

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5 # Probability of success
size = 1000 # Number of samples to generate

# Generate random numbers from a binomial distribution
binomial_distribution = np.random.binomial(n, p, size)

# Plot the histogram
plt.hist(binomial_distribution, bins=np.arange(-0.5, n + 1.5, 1), edgecolor='black', density=True)
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(np.arange(0, n + 1, 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

# **Explanation:**

* np.random.binomial(n, p, size) ‚Üí generates random values from a Binomial distribution.

* n = 10 ‚Üí number of trials.

* p = 0.5 ‚Üí probability of success for each trial.

* size = 1000 ‚Üí number of simulated experiments.

* The histogram shows how often each possible number of ‚Äúsuccesses‚Äù (0 to 10) occurs.

# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# (Include your Python code and output in the code box below).

Answer:
We want to test whether the sample mean is significantly different from the population mean (Œº‚ÇÄ = 50).

    >** Let‚Äôs assume:**

* Population mean (Œº‚ÇÄ) = 50
* Population standard deviation (œÉ) = 0.5
* Significance level (Œ±) = 0.05


‚úÖ Step 1: State the Hypotheses

* Null Hypothesis (H‚ÇÄ): Œº = 50 ‚Üí (No difference from population mean)

* Alternative Hypothesis (H‚ÇÅ): Œº ‚â† 50 ‚Üí (There is a significant difference)



‚úÖ Step 2: Python Code













```python
import numpy as np
from scipy import stats

# Given data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known parameters
mu_0 = 50      # Population mean (claimed)
sigma = 0.5    # Population standard deviation
alpha = 0.05   # Significance level

# Calculate sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Compute Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed critical value for Œ± = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

# p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Display results
print(f"Sample Mean: {sample_mean:.3f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"Critical Z-value: ¬±{z_critical:.3f}")
print(f"P-value: {p_value:.4f}")

# Decision
if abs(z_stat) > z_critical:
    print("üî¥ Reject the Null Hypothesis (H‚ÇÄ)")
else:
    print("üü¢ Fail to Reject the Null Hypothesis (H‚ÇÄ)")
```

# *Step 4: Interpretation*

* The Z-statistic (‚âà 0.88) is less than the critical value (¬±1.96).

* The p-value (‚âà 0.38) is greater than 0.05.

‚û°Ô∏è Therefore, we fail to reject H‚ÇÄ.

# *‚úÖ Conclusion:*

There is no significant difference between the sample mean and the population mean (50).
The sample data is consistent with the population claim.

# **Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

Answer:



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(0)             # for reproducibility
mean = 50                     # true mean
std_dev = 5                   # true standard deviation
n = 100                       # sample size

data = np.random.normal(mean, std_dev, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)   # sample standard deviation

# Step 3: Calculate 95% confidence interval for the mean
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)   # t critical value

margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot histogram with mean and CI
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='dotted', linewidth=2, label=f'Lower 95% CI = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='dotted', linewidth=2, label=f'Upper 95% CI = {ci_upper:.2f}')

plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

# **Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean?**

Answer:


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define a function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores, mean, std

# Step 2: Create a sample dataset
np.random.seed(0)
data = np.random.normal(100, 15, 200)   # Normal distribution: mean=100, std=15, n=200

# Step 3: Call the function
z_scores, mean, std = calculate_z_scores(data)

# Step 4: Display first few Z-scores
print("Original Mean:", round(mean, 2))
print("Original Std Dev:", round(std, 2))
print("First 5 Z-scores:", np.round(z_scores[:5], 2))

# Step 5: Plot histogram of Z-scores
plt.hist(z_scores, bins=15, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
plt.title('Histogram of Z-scores (Standardized Data)')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

# **Interpretation: What Z-scores Represent**

* A Z-score tells how many standard deviations a data point is from the mean.

| Z-score | Meaning                                 |
| ------- | --------------------------------------- |
| Z = 0   | Data point = Mean                       |
| Z = +1  | 1 standard deviation **above** the mean |
| Z = -1  | 1 standard deviation **below** the mean |
| Z = +2  | 2 SDs above (very high value)           |
| Z = -2  | 2 SDs below (very low value)            |
